## Pandas DataFrame and Indexing Method
### Example 1:
##### `DataFrame()`: Used to create a Pandas DataFrame.

````python
data={
    'Name':['Alice','Bob','Charlie','David'],
    'Age':[25,30,22,35],
    'City':['New York','Paris','London','Berlin'],
    'Salary':[70000,80000,60000,90000]
}
df=pd.DataFrame(data)
df
````

##### `loc[row_condition, column_name]`: Used for label-based row and column selection

````python
subset_loc=df.loc[1:3,['Name','Salary']]
subset_loc
````

##### `iloc[row_index, column_index]`: Used for integer position-based row and column selection.

````python
subset_iloc=df.iloc[0:2,1:3]
subset_iloc
````

##### `df[df['column'] condition]`: Used to filter rows based on a condition.

##### Conditions:

````python
high_salary=df[df['Salary']>75000]
high_salary
````

````python
df.loc[df['Age']<30,'Salary']=65000
df
````

## Pandas String Cleaning in DataFrame
### Example 2:
##### `.str.strip()`: Removes leading and trailing spaces from strings.
##### `.str.lower()`: Converts all text to lowercase.
##### `.str.replace('-', ' ')`: Replaces `-` with space in string values.

````python
df=pd.DataFrame({'Category':['Electronics ','furniture','Home-Appliances',' electronics','FURNITURE']})
df
````

````python
df['Category_clean']=df['Category'].str.strip().str.lower()
df
````

````python
df['Category_clean']=df['Category_clean'].str.replace('-',' ')
df
````

## Pandas Merge & Concatenation
### Example 3:

````python
customers=pd.DataFrame({'CustomerID':[1,2,3],'':['Alice','Bob','Charlie']})
orders=pd.DataFrame({'OrderID':[101,102,103,104],'CustomerID':[1,2,2,3],'Amount':[250,150,200,300]})
````

````python
customers
````

````python
orders
````

##### `pd.merge(..., how='inner')`: Used to combine two DataFrames and keep only matching rows from both tables.
##### `pd.merge(..., how='left')`: Used to keep all rows from left DataFrame and matching rows from right.
##### `pd.merge(..., how='right')`: Used to keep all rows from right DataFrame and matching rows from left.

````python
df_inner=pd.merge(customers,orders,on='CustomerID',how='inner')
df_inner
````

````python
df_left=pd.merge(customers,orders,on='CustomerID',how='left')
df_left
````

````python
df_right=pd.merge(customers,orders,on='CustomerID',how='right')
df_right
````

##### `pd.concat()`: Used to join DataFrames vertically or horizontally.

````python
df_concat=pd.concat([customers,customers])
df_concat
````
