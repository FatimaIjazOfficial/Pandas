# Categorical Data Encoding in Python (Pandas + Scikit-Learn)

````python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
````

----
## Categorical Encoding?

Machine Learning models work with **numbers**, not text labels.

So when a dataset contains categories like:

* Cities → Lahore, Karachi, Islamabad
* Sizes → Small, Medium, Large

we must convert them into numerical form.
This process is called **Categorical Encoding**.

---

# Types of Encoding

| Encoding Type    | Used For                | Example                |
| ---------------- | ----------------------- | ---------------------- |
| One-Hot Encoding | Nominal data (no order) | City names             |
| Ordinal Encoding | Ordered categories      | Small < Medium < Large |


----
### One-Hot Encoding
----
One-Hot Encoding converts each category into a separate column.

| City    | -> | City_Lahore | City_Karachi |
| ------- | ---| ----------- | ------------ |
| Lahore  | -> | 1           | 0            |
| Karachi | -> | 0           | 1            |

`drop_first=True`: 
 - removes the first category column from each feature. use in especially in Linear Regression.
 - Why Remove One Column?: To avoid Dummy Variable Trap
 - This happens when columns become highly correlated. Machine learning models may face problems because one column can be predicted from others.

Instead of:

| City_Islamabad | City_Karachi | City_Lahore |
| -------------- | ------------ | ----------- |

it removes one column:

| City_Karachi | City_Lahore |
| ------------ | ----------- |

because the removed category can already be understood.

````python
df = pd.DataFrame({
    "City": ["Lahore", "Karachi", "Islamabad", "Lahore", "Karachi"],
    "Segment": ["A", "B", "C", "D", "E"],
    "Sales": [200, 240, 180, 300, 220]
})

df
````

````python
# Apply One-Hot Encoding

encoded = pd.get_dummies(
    df,
    columns=["City", "Segment"],
    drop_first=True
)

encoded
````

----
### Ordinal Encoding
----

Ordinal Encoding converts categories into numbers **based on order/ranking**.

| Size   | -> | Size   | Encoded |
| ------ | -- | ------ | ------- |
| Small  | -> | Small  | 0       |
| Medium | -> | Medium | 1       |
| Large  | -> | Large  | 2       |

`categories`: defines the order manually.

#### Why Manual Order is Important?

Without defining categories manually:
`OrdinalEncoder()`: may assign alphabetical order which is logically wrong for size comparison:

```python
Large  -> 0
Medium -> 1
Small  -> 2
```

````python
df2 = pd.DataFrame({
    "Size": ["Small", "Medium", "Large", "Medium", "Large"]
})

df2
````

````python
# Apply Ordinal Encoding
encoder = OrdinalEncoder(
    categories=[["Small", "Medium", "Large"]]
)

df2["Size_encoded"] = encoder.fit_transform(df2[["Size"]])

df2
````
